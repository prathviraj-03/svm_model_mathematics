```python
import math
import pandas as pd

# ----------------------------
# Dataset
# ----------------------------
data = pd.DataFrame({
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Weak'],
    'Played': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes']
})

# ----------------------------
# Function: Entropy
# ----------------------------
def entropy(target_col):
    elements, counts = pd.Series(target_col).value_counts().index, pd.Series(target_col).value_counts().values
    e = 0
    for i in range(len(elements)):
        p_i = counts[i]/sum(counts)
        e -= p_i * math.log2(p_i)
    return e

# ----------------------------
# Function: Information Gain
# ----------------------------
def info_gain(data, split_attribute, target_name="Played"):
    total_entropy = entropy(data[target_name])
    
    vals, counts = pd.Series(data[split_attribute]).value_counts().index, pd.Series(data[split_attribute]).value_counts().values
    weighted_entropy = 0
    for i in range(len(vals)):
        subset = data[data[split_attribute] == vals[i]]
        weighted_entropy += (counts[i]/sum(counts)) * entropy(subset[target_name])
    
    IG = total_entropy - weighted_entropy
    return IG

# ----------------------------
# Calculate IG for all attributes
# ----------------------------
attributes = ['Outlook', 'Temperature', 'Humidity', 'Wind']
IG_values = {}

print("Information Gain for each attribute:")
for attr in attributes:
    IG = info_gain(data, attr)
    IG_values[attr] = IG
    print(f"{attr}: {IG:.3f}")

# ----------------------------
# Select Root Attribute (highest IG)
# ----------------------------
root = max(IG_values, key=IG_values.get)
print(f"\nRoot Attribute (Highest IG): {root}")

# ----------------------------
# Simple Tree Diagram (Text)
# ----------------------------
def print_tree(data, target_name, root_attr):
    print(f"\nDecision Tree Root: {root_attr}\n")
    root_vals = data[root_attr].unique()
    for val in root_vals:
        subset = data[data[root_attr] == val]
        e = entropy(subset[target_name])
        print(f"{root_attr} = {val} -> Entropy({target_name}) = {e:.3f}, Samples = {len(subset)}")
        print("   Classes:", subset[target_name].values)
        print("-" * 50)

print_tree(data, 'Played', root)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Sample Data
# ----------------------------
data = np.array([
    [1, 1],
    [1.5, 2],
    [3, 4],
    [5, 7],
    [3.5, 5],
    [4.5, 5],
    [3.5, 4.5]
])

k = 2  # number of clusters
centroids = data[np.random.choice(len(data), k, replace=False)]  # random centroids

# ----------------------------
# Step 2: Distance + Assignment
# ----------------------------
for i in range(5):  # run for fixed iterations
    # Assign points to nearest centroid
    distances = np.array([[np.linalg.norm(x - c) for c in centroids] for x in data])
    cluster_labels = np.argmin(distances, axis=1)

    # Update centroids (mean of cluster points)
    new_centroids = np.array([data[cluster_labels == j].mean(axis=0) for j in range(k)])

    print(f"Iteration {i+1}:")
    print("Centroids:\n", new_centroids, "\n")

    if np.all(centroids == new_centroids):
        break
    centroids = new_centroids

# ----------------------------
# Step 3: Show Results
# ----------------------------
colors = ['red', 'blue']
for i in range(k):
    plt.scatter(data[cluster_labels == i, 0], data[cluster_labels == i, 1], color=colors[i], label=f'Cluster {i+1}')
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=200, label='Centroids')
plt.title("K-Means Clustering (Simple Example)")
plt.legend()
plt.show()
```
