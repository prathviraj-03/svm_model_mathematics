### 2Ô∏è‚É£ üìä Load and Inspect the Dataset
We load the dataset from the specified Excel file and display its shape and the first few rows to understand its structure.

```python
# Change path if needed
dataset_path = r"C:\Users\achar\AppData\Local\Programs\Python\Python311\Scripts\SVM\ASSETS\rebuilt_math_dataset.xlsx"

# sanity check
if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset not found at {dataset_path}. Please verify the path.")

df = pd.read_excel(dataset_path)
print("Dataset shape:", df.shape)
display(df.head())
```

```python
# Show basic info
display(df.info())
display(df.describe(include='all').T)

# Show counts of Performance Level if present
if 'Performance Level' in df.columns:
    print("\nPerformance Level value counts:")
    display(df['Performance Level'].value_counts())
else:
    print("\nNo 'Performance Level' column found - please add it then re-run.")
```

```python
# Mapping for Likert responses
likert_map = {
    "Very Poor": 1,
    "Poor": 2,
    "Average": 3,
    "Good": 4,
    "Excellent": 5
}

# Create a copy so we don't overwrite original unexpectedly
data = df.copy()

# Apply mapping to all object dtype columns except target
target_col = "Performance Level"

for col in data.columns:
    if col == target_col:
        continue
    if data[col].dtype == object:
        # replace known strings, keep other strings (like numeric-as-strings) handled by to_numeric
        data[col] = data[col].replace(likert_map)

# Try casting numeric columns
for col in data.columns:
    if data[col].dtype == object:
        # attempt to convert to numeric if they are numeric-like strings
        data[col] = pd.to_numeric(data[col], errors='ignore')

print("After mapping ‚Äî sample:")
display(data.head())
```

```python
# If target is textual (Low/Medium/High), encode them; if numeric already, keep as-is
if target_col not in data.columns:
    raise KeyError(f"Target column '{target_col}' not found. Add it and re-run.")

# If target is string labels like Low/Medium/High, encode
if data[target_col].dtype == object:
    le = LabelEncoder()
    data[target_col] = le.fit_transform(data[target_col])
    print("Encoded target classes:", dict(enumerate(le.classes_)))
else:
    # If numeric but possibly floats, round/convert to int
    if pd.api.types.is_float_dtype(data[target_col].dtype):
        # if your target is already a score (1-5) or categories (1/2/3) use as is
        data[target_col] = data[target_col].round().astype(int)

# Check for missing values and fill numeric columns with mean
if data.isna().sum().sum() > 0:
    print("Missing values found ‚Äî imputing numeric columns with mean")
    numeric_cols = data.select_dtypes(include=[np.number]).columns
    data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

display(data.head())
print("\nTarget value counts:")
display(data[target_col].value_counts())
```

```python
# Drop columns that are not features if any (like student ids) - attempt to auto-detect common names
exclude = [target_col]
# If columns such as "Timestamp" or "Email" exist, drop them too
for typical in ['Timestamp', 'Email', 'Name', 'StudentID', 'ID']:
    if typical in data.columns:
        exclude.append(typical)

X = data.drop(columns=exclude)
y = data[target_col]

print("Features shape:", X.shape)
print("Target shape:", y.shape)
display(X.columns)
```

```python
# Use stratify to keep class distribution
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Optionally, put back into DataFrame for readability
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled, columns=X.columns, index=X_test.index)
```

```python
# Baseline SVM (RBF)
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=42)
svm_rbf.fit(X_train_scaled, y_train)

y_pred = svm_rbf.predict(X_test_scaled)

acc = accuracy_score(y_test, y_pred)
print(f"Baseline SVM (RBF) Accuracy: {acc*100:.2f}%")
print("\nClassification report:")
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - SVM (RBF)")
plt.show()
```

```python
svm_lin = SVC(kernel='linear', C=1.0, probability=True, random_state=42)
svm_lin.fit(X_train_scaled, y_train)
y_pred_lin = svm_lin.predict(X_test_scaled)

acc_lin = accuracy_score(y_test, y_pred_lin)
print(f"Linear SVM Accuracy: {acc_lin*100:.2f}%")
print(classification_report(y_test, y_pred_lin))
```
